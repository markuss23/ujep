```python

import numpy as np  
import matplotlib.pyplot as plt  
def generate_fm_signal(carrier_freq, modulating_freq, modulation_index, duration, sampling_rate):
    """
    Generuje frekvenn modulovan (FM) signl.

    Args:
        carrier_freq (float): Frekvence nosnho signlu v Hz.
        modulating_freq (float): Frekvence modulanho signlu v Hz.
        modulation_index (float): Index modulace.
        duration (float): Dlka signlu v sekundch.
        sampling_rate (float): Vzorkovac frekvence v Hz.

    Returns:
        tuple: asov osa a odpovdajc hodnoty FM signlu.
    """
    # asov osa
    t = np.linspace(0, duration, int(sampling_rate * duration))

    # Modulan signl (nap. sinusov vlna)
    modulating_signal = np.sin(2 * np.pi * modulating_freq * t)

    # FM signl (nosn signl modulovan frekvenc)
    fm_signal = np.cos(2 * np.pi * carrier_freq * t + modulation_index * modulating_signal)

    return t, fm_signal



# Parametry signlu
carrier_freq = 10  # Frekvence nosnho signlu v Hz
modulating_freq = 1  # Frekvence modulanho signlu v Hz
modulation_index = 5  # Index modulace
duration = 5  # Dlka signlu v sekundch
sampling_rate = 1000  # Vzorkovac frekvence v Hz

# Generovn FM signlu
t, fm_signal = generate_fm_signal(carrier_freq, modulating_freq, modulation_index, duration, sampling_rate)

# Vykreslen FM signlu
plt.figure(figsize=(10, 4))
plt.plot(t, fm_signal)
plt.title('Frekvenn modulovan (FM) signl')
plt.xlabel('as [s]')
plt.ylabel('Amplituda')
plt.grid(True)

# Zobrazen grafu
plt.show()

```

```python
import numpy as np  
import matplotlib.pyplot as plt  
  
# Sampling frequency  
fs = 1000
# Carrier frequency  
fc = 10
# Modulation frequency  
fm = 1  
# Modulation index  
beta = 5  
# Time  
t = np.arange(0, 1, 1/fs)  
  
# Modulating signal  
x = np.sin(2*np.pi*fm*t)  
  
# Carrier signal  
c = np.sin(2*np.pi*fc*t)  
  
# Frequency modulation  
y = np.sin(2*np.pi*(fc + beta*x)*t)  
  
# Plot the signals  
plt.figure(figsize=(10, 6))  
plt.plot(t, x, label='Modulating signal')  
plt.plot(t, c, label='Carrier signal')  
plt.plot(t, y, label='FM signal')  
plt.legend()  
plt.xlabel('Time (s)')  
plt.ylabel('Amplitude')  
plt.show()  
```
