```python
import numpy as np
import matplotlib.pyplot as plt

# Parametry nosné vlny
frekvence_nosne = 100  # Hz
amplituda_nosne = 1
cas = np.linspace(0, 1, 1000)  # 1 sekunda, 1000 vzorků

# Vytvoření nosné vlny
nosna_vlna = amplituda_nosne * np.sin(2 * np.pi * frekvence_nosne * cas)

# Parametry modulačního signálu
frekvence_modulacni = 5  # Hz
amplituda_modulacni = 0.5

# Vytvoření modulačního signálu
modulacni_signal = amplituda_modulacni * np.sin(2 * np.pi * frekvence_modulacni * cas)

# Vytvoření modulované vlny (AM modulace)
modulovana_vlna = (1 + modulacni_signal) * nosna_vlna

# Vykreslení
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(cas, nosna_vlna)
plt.title('Nosná vlna')
plt.ylabel('Amplituda')

plt.subplot(3, 1, 2)
plt.plot(cas, modulacni_signal)
plt.title('Modulační signál')
plt.ylabel('Amplituda')

plt.subplot(3, 1, 3)
plt.plot(cas, modulovana_vlna)
plt.title('Modulovaná vlna (AM)')
plt.xlabel('Čas (s)')
plt.ylabel('Amplituda')

plt.tight_layout()
plt.show()
```
