```python
import numpy as np
import matplotlib.pyplot as plt

# Parametry nosn vlny
frekvence_nosne = 100  # Hz
amplituda_nosne = 1
cas = np.linspace(0, 1, 1000)  # 1 sekunda, 1000 vzork

# Vytvoen nosn vlny
nosna_vlna = amplituda_nosne * np.sin(2 * np.pi * frekvence_nosne * cas)

# Parametry modulanho signlu
frekvence_modulacni = 5  # Hz
amplituda_modulacni = 0.5

# Vytvoen modulanho signlu
modulacni_signal = amplituda_modulacni * np.sin(2 * np.pi * frekvence_modulacni * cas)

# Vytvoen modulovan vlny (AM modulace)
modulovana_vlna = (1 + modulacni_signal) * nosna_vlna

# Vykreslen
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(cas, nosna_vlna)
plt.title('Nosn vlna')
plt.ylabel('Amplituda')

plt.subplot(3, 1, 2)
plt.plot(cas, modulacni_signal)
plt.title('Modulan signl')
plt.ylabel('Amplituda')

plt.subplot(3, 1, 3)
plt.plot(cas, modulovana_vlna)
plt.title('Modulovan vlna (AM)')
plt.xlabel('as (s)')
plt.ylabel('Amplituda')

plt.tight_layout()
plt.show()
```
