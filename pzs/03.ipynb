```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for two linear lines
x = np.linspace(-10, 10, 100)
# y1 =1 * x + 1
# Generate a rectangle wave signal
y1 = np.sign(np.sin(x)*2)
# y2 = 1 * x**2 - 2
# Generate a sine wave signal
y2 = np.sin(x)
def get_covariance(x, y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    covariance = np.sum((x - x_mean) * (y - y_mean)) / len(x)
    return covariance

# Calculate the covariance matrix
covariance_matrix = np.cov(y1, y2)
# Extract the covariance value
covariance_value = covariance_matrix[0, 1]  # or covariance_matrix[1, 0]

covariance = get_covariance(y1, y2)
print(covariance)


# Plot the lines
plt.plot(x, y1, label='y1 = 2x + 1', color='blue')
plt.plot(x, y2, label='y2 = 3x + 10', color='orange')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Two Linear Lines')

# Show the covariance value on the plot
plt.text(1, 15, f'Covariance: {covariance:.2f}', fontsize=12, color='black')

# Add a legend
plt.legend()

# Show the plot
plt.show()

# Print the covariance value
print(f'Covariance between y1 and y2: {covariance:.2f}')
```
