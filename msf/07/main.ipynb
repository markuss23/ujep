```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x, y):
    
    return x**2/2 + 2

# Set initial conditions
x0 = 0
y0 = 2
h = 1  # step size
x_end = 10

# Initialize arrays to store the values of x and y
x_values = [x0]
y_values = [y0]

while x_values[-1] < x_end:
    current_x = x_values[-1]
    current_y = y_values[-1]
    next_y = current_y + h * f(current_x, current_y)
    next_x = current_x + h
    x_values.append(next_x)
    y_values.append(next_y)

# Plot the solution
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label="y(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin
# Define the function
def f(x, y):
    return sin(x)+1

# Set initial conditions
x0 = 0
y0 = 2
h = 0.01  # step size
x_end = 10

# Initialize arrays to store the values of x and y
x_values = np.array([x0])
y_values = np.array([y0])

while x_values[-1] < x_end:
    current_x = x_values[-1]
    current_y = y_values[-1]
    next_y = current_y + h * f(current_x, current_y)
    next_x = current_x + h
    x_values = np.append(x_values, next_x)
    y_values = np.append(y_values, next_y)

# Plot the solution
plt.figure(figsize=(10, 5))
plt.plot(x_values, y_values, label="y(x)")
plt.legend()
plt.grid(True)
plt.show()
```
