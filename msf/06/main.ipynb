```python
import numpy as np
from matplotlib import pyplot as plt
def f(x):
    """_summary_

    Args:
        x (_type_): _description_

    Returns:
        _type_: _description_
    """
    return -x**3 + 10*x**2-3*x+6

def linear(x,a,b):
    return f(a) + (f(b)-f(a))/(b-a)*(x-a)

def vandermond_matrix(x):
    return np.vander(x,increasing=True)
    

xstart = 0
xend = 10

nx = 5

dx = (xend-xstart)/nx
x = np.linspace(xstart,xend,nx)
fx = f(x)


phi = []
xphi = []
ninterpol = 5
for i in range(len(x)-1):
    xinterpol = np.linspace(x[i],x[i+1],ninterpol)
    g = linear(xinterpol,x[i],x[i+1])
    xphi.extend(xinterpol)
    phi.extend(g)

# x = np.arange(xstart,xend)
# y = np.random.randint(0,10,len(x))
# X = vandermond_matrix(x)
# print(X)
# a = np.linalg.solve(X,y)

plt.plot(xphi,phi,"go-")
plt.plot(x,fx, "ro")
```
