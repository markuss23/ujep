```python
import numpy as np

def generate_matrix(N):
    return np.random.randn(N, N)

def generate_vector(N):
    return np.random.randn(N)

def is_matrix_solvable(matrix):
    det = np.linalg.det(matrix)
    return det != 0

def jacobi_iteration(A, b, initial_guess, tolerance=1e-10, max_iterations=1000):
    n = len(b)
    x = initial_guess.copy()
    D = np.diag(np.diag(A))
    LU = A - D

    for _ in range(max_iterations):
        x_new = np.dot(np.linalg.inv(D), b - np.dot(LU, x))
        if np.linalg.norm(x_new - x) < tolerance:
            return x_new
        x = x_new

    raise ValueError("Jacobi method did not converge within the maximum number of iterations.")    

def jacobi(A, b, tolerance=1e-10, max_iterations=10000):
    
    x = np.zeros_like(b, dtype=np.double)
    
    T = A - np.diag(np.diagonal(A))
    
    for k in range(max_iterations):
        
        x_old  = x.copy()
        
        x[:] = (b - np.dot(T, x)) / np.diagonal(A)
        
        if np.linalg.norm(x - x_old, ord=np.inf) / np.linalg.norm(x, ord=np.inf) < tolerance:
            break
            
    return x

def gauss_seidel(A, b, tolerance=1e-10, max_iterations=10000):
    
    x = np.zeros_like(b, dtype=np.double)
    
    #Iterate
    for k in range(max_iterations):
        
        x_old  = x.copy()
        
        #Loop over rows
        for i in range(A.shape[0]):
            x[i] = (b[i] - np.dot(A[i,:i], x[:i]) - np.dot(A[i,(i+1):], x_old[(i+1):])) / A[i ,i]
            
        #Stop condition 
        if np.linalg.norm(x - x_old, ord=np.inf) / np.linalg.norm(x, ord=np.inf) < tolerance:
            break
            
    return x


def sor_method(A, b, initial_guess, omega, tolerance=1e-10, max_iterations=1000):
    """
    Solves the linear system Ax = b using the Successive Over-Relaxation (SOR) method.

    Parameters:
        A: numpy.ndarray
            Coefficient matrix (NxN).
        b: numpy.ndarray
            Right-hand side vector (N).
        initial_guess: numpy.ndarray
            Initial guess for the solution vector (N).
        omega: float
            Relaxation parameter (0 < omega < 2).
        tolerance: float, optional
            Tolerance for convergence.
        max_iterations: int, optional
            Maximum number of iterations.

    Returns:
        numpy.ndarray
            Solution vector (N).
    """
    n = len(b)
    x = initial_guess.copy()

    for iteration in range(max_iterations):
        x_old = x.copy()
        for i in range(n):
            sigma = np.dot(A[i, :i], x[:i]) + np.dot(A[i, i + 1 :], x_old[i + 1 :])
            x[i] = (1 - omega) * x_old[i] + (omega / A[i, i]) * (b[i] - sigma)
        if np.linalg.norm(np.dot(A, x) - b) < tolerance:
            return x

    raise ValueError("SOR method did not converge within the maximum number of iterations.")


N = 100
matrix = generate_matrix(N)
vector = generate_vector(N)

# solvable = is_matrix_solvable(matrix)
# for _ in range(10):
#     m = generate_matrix(1000)
#     if is_matrix_solvable(m):
#         print("Matrix is solvable")
#     else:
#         print("Matrix is not solvable")
    
# if solvable:
#     print("Matrix is solvable")
#     result = np.linalg.solve(matrix, vector)
#     print(result)
# else:
#     print("Matrix is not solvable")

print("Solving using Jacobi method")
A = np.array([[10, 2, 1], [1, 5, 1], [2, 3, 10]])
b = np.array([7, -8, 6])
initial_guess = np.zeros_like(b)
omega = 1.2
solution = jacobi_iteration(A, b, initial_guess)
print("Solution:", solution)
print("asd", jacobi(A, b))
print("Actual solution:", np.linalg.solve(A, b))
print("gauss_seidel", gauss_seidel(A, b))

  
```
