```python
from matplotlib import pyplot as plt
import numpy as np

def my_function(n, max, min):
    x = np.linspace(min, max, n)
    y = np.sin(x)
    return x, y

n = 10
min = 0
max = 2 * np.pi
x, y = my_function(n, max, min)
y2 = y * 2


plt.figure(figsize=(8,8), dpi=80)
plt.subplot(2, 2, 1)  
plt.bar(x, y)  
plt.title('Plot 1')

plt.subplot(2, 2, 2) 
plt.scatter(x, y2)  
plt.title('Plot 2')

plt.subplot(2, 1, 2) 
plt.bar(x, y2)  
plt.barh(x+1, y, alpha=0.8)  





plt.show()  
```
